```python
from netCDF4 import Dataset
from netCDF4 import num2date, date2num
import netCDF4
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.display import Image
import pickle
import numpy as np
import fnmatch
import os
```

```python
fileLoc='/xace/d1/pblh_archive/CLAC_wrfchem/wrfout_d01_2018-01-01_00:00:00'
```

```python
path='/xace/d1/pblh_archive/CLAC_wrfchem/'
fileNameList=[]
allFileList=[]
count=0
pattern=['wrfout_d01*','wrfout_d01_2019-05*','wrfout_d01_2019-06*']
for filename in os.listdir(path):
    allFileList.append(filename)
allFileList.sort()
for filename in allFileList:
    if fnmatch.fnmatch(filename, pattern[1]):
        count+=1
        #if count<=10000:#file for first 40 days
        fileNameList.append(filename)
```

```python

times = []
latitude = np.ma.zeros((29,29))
longitude = np.ma.zeros((29,29))
no2 = np.ma.zeros((29,29,29))
no = np.ma.zeros((29,29,29))
o3 = np.ma.zeros((29,29,29))
hno3 = np.ma.zeros((29,29,29))
h2o2 = np.ma.zeros((29,29,29))
temperature = np.ma.zeros((29,29,29))
xWind = np.ma.zeros((29,29,30))
yWind = np.ma.zeros((29,30,29))
zWind = np.ma.zeros((30,29,29))
pressure = np.ma.zeros((29,29,29))
```

```python
count=0
for ncfile in fileNameList:
    count+=1
    wrf = Dataset('/xace/d1/pblh_archive/CLAC_wrfchem/'+ncfile, "r")
    time  = wrf.variables['Times'][:].squeeze()
    time = ''.join(x.replace("b", "").replace(" ", "").replace("'","").replace("\n","") for x in str(time))
    XLAT = wrf.variables['XLAT'][:].squeeze()
    XLONG = wrf.variables['XLONG'][:].squeeze()
    wrf_no2 = wrf.variables['no2'][:]
    wrf_no = wrf.variables['no'][:]
    wrf_o3 = wrf.variables['o3'][:]
    wrf_hno3 = wrf.variables['hno3'][:]
    wrf_h2o2 = wrf.variables['h2o2'][:]
    T = wrf.variables['T'][:]
    U = wrf.variables['U'][:]
    V = wrf.variables['V'][:]
    W = wrf.variables['W'][:]
    P = wrf.variables['P'][:]
    times.append(time)
    
    if count == 1:
        latitude = XLAT
        longitude = XLONG
        no2 = wrf_no2
        no = wrf_no
        o3 = wrf_o3
        hno3 = wrf_hno3
        h2o2 = wrf_h2o2
        temperature = T
        xWind = U
        yWind = V
        zWind = W
        pressure = P
    else:
        temperature = np.ma.concatenate([temperature,T])
        xWind = np.ma.concatenate([xWind,U])
        yWind = np.ma.concatenate([yWind,V])
        zWind = np.ma.concatenate([zWind,W])
        pressure = np.ma.concatenate([pressure,P])
        no2 = np.ma.concatenate([no2,wrf_no2])
        no = np.ma.concatenate([no,wrf_no])
        o3 = np.ma.concatenate([o3,wrf_o3])
        hno3 = np.ma.concatenate([hno3,wrf_hno3])
        h2o2 = np.ma.concatenate([h2o2,wrf_h2o2])
```

```python
print('times :' + str(len(times)))
print('Lat :' + str(latitude.shape))
print('Lon :' + str(longitude.shape))
print('Temperature :' + str(temperature.shape))
print('xWind :' + str(xWind.shape))
print('yWind :' + str(yWind.shape))
print('zWind :' + str(zWind.shape))
print('pressure :' + str(pressure.shape))
print('no2 :' + str(no2.shape))
print('no :' + str(no.shape))
print('o3 :' + str(o3.shape))
print('hno3 :' + str(hno3.shape))
print('h2o2 :' + str(h2o2.shape))
```

```python
data = {
    "times": times,
    "latitude": latitude,
    "longitude": longitude,
    "temperature": temperature,
    "xWind": xWind,
    "yWind": yWind,
    "zWind": zWind,
    "pressure":pressure,
    "no2": no2,
    "no": no,
    "o3": o3,
    "hno3": hno3,
    "h2o2":h2o2
}
```

```python
datafile = open('PickleDataFiles/wrfout_d01_2019_05_01_to_2019_05_31_TPWHNO', 'wb')
# source, destination 
pickle.dump(data, datafile)                      
datafile.close()
```
